```python

# Read the DataFrame
df1 = pd.read_csv('updated.csv')

Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)

# Calculate the median for each column
median = df1.median()

for column in df1.columns:
    df1.loc[outliers, column] = median[column]

df1.to_csv('updated_data.csv', index=False)
```
